```python
"""" 
2. All the other columns in the dataset are the attributes. 
For each attribute: compute the mean and the standard deviation. 
For all of the attributes together, create a correlation matrix 
(Using Pearson correlation coefficients). Please use numpy and scipy methods. 
Next, calculate a correlation matrix using the pandas method (your dataframe variable name).corr. 
Did you get the same values?
"""
my_corr = []
Y = df["Y"].values
mean_Y = np.mean(Y)
N = len(df)
mean = {}
std_dev = {}
corr = []


for (column_name, _) in df.iteritems():
    X = df[column_name].values
    mean[f"{column_name}"] = np.mean(X)
    std_dev[f"{column_name}"] = (np.sum((X - np.mean(X))**2))/(N-1)

print(f"mean: {mean}")
print(f"standard dev: {std_dev}")
print(np.array(df))

# Pearson product-moment correlation: [-1, +1] -1: 负相关 -> +1: 正相关

print("Correlation coefficient matrix (computed with numpy):") 
# np.corrcoef (Pearson product-moment correlation efficient) 
# https://numpy.org/doc/stable/reference/generated/numpy.corrcoef.html
# 如何理解np.array(df).T：因为这里的研究对象是每一列数据，而原始数据是以row为变量的，所以需要转置
display(pd.DataFrame(np.corrcoef(np.array(df).T),columns=df.columns))

print("Correlation coefficient matrix (computed with pandas):")
display(df.corr())
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

# It does not make sense to use linear regression on the sex column, so we drop it
diabetes = df.drop(columns = ['SEX'])

"""
1. Normalize all columns which it makes sense to normalize

Why normalizing data:

Standardization of a dataset is a common requirement for many machine learning estimators: they 
might behave badly if the individual features do not more or less look like standard normally 
distributed data (e.g. Gaussian with 0 mean and unit variance).

For instance many elements used in the objective function of a learning algorithm (such as the RBF 
kernel of Support Vector Machines or the L1 and L2 regularizers of linear models) assume that all 
features are centered around 0 and have variance in the same order. If a feature has a variance 
that is orders of magnitude larger that others, it might dominate the objective function and make 
the estimator unable to learn from other features correctly as expected.

"""
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
saved_cols = diabetes.columns
# Fits transformer to X and y (none for non-supervised transform) and returns a transformed version of X.
diabetes_arr = scaler.fit_transform(diabetes)
display(pd.DataFrame(diabetes_arr, columns=saved_cols))


"""
2. Separate the entries into train and test segments randomly. Have the train segment contain 80%, about 354 
entries. Let the test segment contain the remaining entries.
"""

from sklearn.model_selection import train_test_split

train, test = train_test_split(diabetes, test_size=0.2)
train_X = train.drop(columns = ["Y"])
train_y = train["Y"]
test_X = test.drop(columns = ["Y"])
test_y = test["Y"]


"""
3. Use scikit-learn's linear regression algorithm to train on two columns of the train set, 
and test on the test set.
"""
from sklearn.linear_model import LinearRegression
diab_pred = LinearRegression().fit(train_X[["AGE", "BP"]], train_y)
print("Result for Age and Blood Pressure", diab_pred.score(test_X[["AGE", "BP"]], test_y))


# Test for all columns
diab_pred = LinearRegression().fit(train_X, train_y)
print("Result for regressing on all columns",diab_pred.score(test_X, test_y))

# the larger the score the better
```
